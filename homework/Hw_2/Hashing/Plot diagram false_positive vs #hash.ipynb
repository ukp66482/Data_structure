```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker

num_hash = range(1,15,1)
false_positive = [0.139233,0.067226,0.047546,0.041418,0.040887,0.043755,0.049151,0.056872,0.067388,0.080532,0.096498,0.115239,0.136997,0.161817] 

min_x = np.argmin(false_positive)+1
min_y = np.min(false_positive)

plt.scatter(min_x,min_y,s=100,color='r',label = "minimum")
lines = plt.plot(num_hash,false_positive)

plt.gca().yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1,decimals=2))


plt.title("false_positive rate vs #hash functions") # title
plt.ylabel("false_positive rate") # y label
plt.xlabel("#hash functions") # x label
plt.setp(lines,marker = ".") 
plt.legend()
```
